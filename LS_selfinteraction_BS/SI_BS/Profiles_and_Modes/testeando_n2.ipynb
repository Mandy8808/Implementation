```python
#########################################
# Calculando los autovalores (Espectro) #
#########################################

def NpuntosCho(s0):
  if s0>=3.5:
    Nptos = int(rMax/2)
  elif s0>=2.5:
    Nptos = int(rMax/4)
  elif s0>=1.5:
    Nptos = int(rMax/5)
  elif s0>=0.5:
    Nptos = int(rMax/6)
  elif s0>=0.05:
    Nptos = int(rMax/7)
  else:
    Nptos = int(rMax/8)
  return Nptos

dist = [0.7, 0.8, 0.9, 1]
ptosVal = [100, 250, 500, 750, 1000]

# Ncasos = range(len(datosProfiles))  # numero de soluciones que se estudiarn

# cada fila es un valor de Nptos y columna una solucin sigma0
Lval = 1

Auto_Real, Auto_Valores, Auto_Funciones = [], [], []
dataPtos = []
for k in Ncasos: #Ncasos:
    fsNtmp, fuNtmp = datosProfiles[k][0], datosProfiles[k][2]

    rfin, alpha, Lamb, nodos = soluciones_Fondo[k][1], soluciones_Fondo[k][2], soluciones_Fondo[k][3], soluciones_Fondo[k][4]
    
    s0 = fsNtmp(0)
    rMax = rfin+200*(nodos+1) #dist[2]*rDnew[-1]  #rfin+200*(nodos+1)
    Nptos = NpuntosCho(s0)+1000 #NpuntosCho(s0)+ptosVal[3] # NpuntosCho(s0)+1000
    dataPtos.append([s0, rMax, Nptos])
    print(Nptos, fsNtmp(rMax), Edata[k][1]-fuNtmp(rMax))
    
    L = Lval
    util = [Nptos, rMax, Lamb, alpha, L]    
    datFunc = [fsNtmp, fuNtmp]

    lambd, _, Vecto_Full, _ = ru.espectroL(datFunc, util, inf=False)

    #### Separando los Autovalores ######
    jj = np.real(lambd)!=0  # reales
    autoval_Real = lambd[jj]
    print('')
    print('Autovalores con part. Real para rho_0=%7.6f, Npt.=%d -> '%(fsNtmp(0), Nptos), autoval_Real)
    
    if len(autoval_Real)!=0:
        #temp = len(autoval_Real)
        #name = 'data_real%d.dat'%(k)
        #data = np.array([[k]*temp, autoval_Real.real, autoval_Real.imag]).T
        #np.savetxt(name, data)
        Auto_Real.append(k)
    ###############

    ###### Organizando de menor a mayor los autovalores #####
    indF = ru.menor(lambd).astype(int)
    autova_Organ = lambd[indF]

    ###### Salvando  #####
    Auto_Valores.append(autova_Organ)
    Auto_Funciones.append(Vecto_Full[:, indF])
```
